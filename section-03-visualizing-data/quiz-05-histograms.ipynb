## Quiz 5: Histograms

A company sells bags of candy. The bags are supposed to weigh approximately 1 pound each, but sometimes they weigh more or less. The company wants to check the weights of their bags to make sure they are consistent and not losing money by putting too much candy in the bags. It's not an issue if the bags weigh less than a pound, but if they weigh too much, then the company begins to lose profit on the bags because of too much candy. They weigh a sample of 200 bags and create a histogram to show the data. What does the shape of the histogram suggest about the weights of the bags?



Most of the candies are close to a pound. Some are just slightly less.
